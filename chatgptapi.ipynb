```python

def classify_intent(text):

        completion = openai.ChatCompletion.create(
                model="gpt-3.5-turbo", # this is "ChatGPT" $0.002 per 1k tokens
                messages=[{"role": "user", "content": "Classify " f"{text}" "into Greet, Goodbye, Inquiry, Offer_price, Disagree, Agree, Vague-price and only give the classification name"},
                        ],
                n=1
        
        )
        user_intent = completion.choices[0].message.content

        return user_intent
```

```python
def response(user_input):

    # intent = self.get_intent(user_input)

    # user_price = self.price_extraction(user_input)

    message_history = [{"role": "system", "content": "You are a online seller who sells switch, the switch you bought one year ago and you wanna a new version of it, so you decided to sell it."}]

    # if intent == self.counter_price:

    #     prompt = intent(user_price)
    
    # else:

    #     prompt = intent()
    
    message_history.append(
        {"role": "user", "content": f"{user_input}"}
        # {"role": "system", "content": reply_content}
        )

    completion = openai.ChatCompletion.create(
        model="gpt-3.5-turbo", # this is "ChatGPT" $0.002 per 1k tokens
        messages= message_history,
        temperature = 0.3,
        n=1,
    )
    reply_content = completion.choices[0].message.content
    message_history.append(
        {"role": "system", "content": f"{reply_content}"}
    )

    return reply_content
```
